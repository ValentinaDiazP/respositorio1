```python
from datetime import datetime

class Implantes:
    def __init__(self, fechaImplantacion, medico, estado, material, IDpaciente):
        self.__fechaImplantacion= fechaImplantacion
        self.__medico= medico
        self._estado= estado
        self.material= material
        self.__IDpaciente = IDpaciente #Para asociar el ID con el implante
        self.__revision= []    
                   
    @property
    def fechaImplantacion(self):
        return self.__fechaImplantacion
    @fechaImplantacion.setter
    def fechaImplantacion(self, fechaImplantacionNueva):
        self.__fechaImplantacion= fechaImplantacionNueva

    @property
    def medico(self):
        return self.__medico
    @medico.setter
    def medico(self, medicoNuevo):
        self.__medico= medicoNuevo
    @property
    def estado(self):
        return self._estado
    @estado.setter
    def estado(self, estadoNuevo):
        self._estado= estadoNuevo

    @property
    def IDpaciente(self):
        return self.__IDpaciente
    @IDpaciente.setter
    def IDpaciente(self, IDpacienteNuevo):
        self.__IDpaciente=IDpacienteNuevo

    def agregar_revision(self, fecha_revision):
        self.__revision.append(fecha_revision)
    def mostrar_revision(self):
        if self.__revision:
            print("Fechas de revisión: ")
            for i in self.__revision:
                print(i)
        else:
            print("No hay revisiones registradas. ")
    
class Marcapasos(Implantes):
    def __init__(self, fechaImplantacion, medico, estado, material, IDpaciente, electrodos, sistema, frecuencia_estimulacion):
        super().__init__(fechaImplantacion, medico, estado, material, IDpaciente)
        self.electrodos=electrodos
        self.sistema=sistema
        self.frecuencia_estimulacion=frecuencia_estimulacion

class stentCoronario(Implantes):
    def __init__(self, fechaImplantacion, medico, estado, material, IDpaciente, longitud, diametro):
        super().__init__(fechaImplantacion, medico, estado, material, IDpaciente)
        self.longitud= longitud
        self.diametro= diametro

class implanteDental(Implantes):
    def __init__(self, fechaImplantacion, medico, estado, material, IDpaciente, fijacion, forma):
        super().__init__(fechaImplantacion, medico, estado, material, IDpaciente)
        self.fijacion= fijacion
        self.forma=forma

class implanteRodilla(Implantes):
    def __init__(self, fechaImplantacion, medico, estado, material, IDpaciente, fijacion, tamanoRodilla):
        super().__init__(fechaImplantacion, medico, estado, material, IDpaciente)
        self.fijacion= fijacion
        self.tamanoRodilla=tamanoRodilla

class implanteCadera(Implantes):
    def __init__(self, fechaImplantacion, medico, estado, material, IDpaciente, fijacion, tamanoCadera):
        super().__init__(fechaImplantacion, medico, estado, material, IDpaciente)
        self.fijacion= fijacion
        self.tamanoCadera=tamanoCadera

class Paciente:
    def __init__(self, nombre, edad, sexo):
        self.nombre= nombre
        self.edad= edad
        self.sexo=sexo
        self.__implantes=[]
        
    def asignar_implante(self, fechaImplantacion, medico, estado, material, IDpaciente, electrodos=None, sistema=None, frecuencia_estimulacion=None, longitud=None, diametro=None, fijacion=None, forma=None, tamanoRodilla=None, tamanoCadera=None):
        if isinstance(fechaImplantacion, datetime):
            if isinstance(medico, str) and isinstance(estado, str) and isinstance(material, str) and isinstance(IDpaciente, int):
                
                if isinstance(electrodos, int) and isinstance(sistema, str) and isinstance(frecuencia_estimulacion, int):
                    implante = Marcapasos(fechaImplantacion, medico, estado, material, IDpaciente, electrodos, sistema, frecuencia_estimulacion)
                elif isinstance(longitud, int) and isinstance(diametro, int) and isinstance(material, str):
                    implante = stentCoronario(fechaImplantacion, medico, estado, material, IDpaciente, longitud, diametro)
                elif isinstance(forma, str) and isinstance(fijacion, str):
                    implante = implanteDental(fechaImplantacion, medico, estado, material, IDpaciente, fijacion, forma)
                elif isinstance(tamanoRodilla, str) and isinstance(fijacion, str) and isinstance(material, str):
                    implante = implanteRodilla(fechaImplantacion, medico, estado, material, IDpaciente, fijacion, tamanoRodilla)
                elif isinstance(tamanoCadera, str) and isinstance(fijacion, str) and isinstance(material, str):
                    implante = implanteCadera(fechaImplantacion, medico, estado, material, IDpaciente, fijacion, tamanoCadera)
                else:
                    print("Error: Los tipos de datos de los atributos del implante no son válidos.")
                
                self.__implantes.append(implante)
            else:
                print("Error: Los tipos de datos de los atributos de fecha, médico, estado y material no son válidos.")
                
        else:
            print("Error: El tipo de datos de la fecha de implantación no es válido.")
            


    def mostrar_implantes(self):                
        for i in self.__implantes:
            print(f"Fecha de implantación: {i.fechaImplantacion}")
            print(f"Médico responsable: {i.medico}")
            print(f"Estado del implante: {i.estado}")
            if isinstance(i, Marcapasos):                          #insistance para comprobar si un objeto es de una instancia de una clase
                print(f"""Número de electrodos: {i.electrodos}
                          Alamabrico o inalámbrico: {i.sistema}
                          Frecuencia: {i.frecuencia_estimulacion}        
                      """)
            elif isinstance(i,stentCoronario):
                print(f"""longitud: {i.longitud}
                          Diámetro: {i.diametro}
                          Material: {i.material}        
                      """)
            elif isinstance(i, implanteDental):
                print(f"""Fijación: {i.fijacion}")
                          Forma: {i.forma}
                    """)
            elif isinstance(i, implanteRodilla):
                print(f"""Fijación: {i.fijacion}")
                          Tamaño de rodilla: {i.tamanoRodilla}
                    """)
            elif isinstance(i, implanteCadera):
                print(f"""Fijación: {i.fijacion}
                          Tamaño de cadera: {i.tamanoCadera}
                    """)                     
            i.mostrar_revision()
    def eliminar_implante(self, ID_paciente):
        implante_encontrado = False
        for c in self.__implantes:
            if c.IDpaciente == ID_paciente:
                self.__implantes.remove(c)
                print("Implante eliminado del paciente.")
                implante_encontrado = True
                break
        if not implante_encontrado:
            print("Implante no encontrado en el paciente.")
    
    def editar_implante(self, IDimplante=None, fechaImplantacion=None, medico=None, estado=None, material=None,**kwargs): #**kwargs para crear un diccionario, la clave es el atributo, la llave el valor pasado
        implante_encontrado = False
        for implante in self.__implantes:
            if implante.IDpaciente == IDimplante:
                implante_encontrado = True
                if fechaImplantacion is not None:
                    implante.fechaImplantacion = fechaImplantacion
                if medico is not None:
                    implante.medico = medico
                if estado is not None:
                    implante.estado = estado
                if material is not None:
                    implante.material= material
                for key, value in kwargs.items():
                    setattr(implante, key, value)             #Establece el valor de un atributo de un objeto (objeto atributo, nombre atributo, valor atributo)
                print("Implante editado correctamente.")
                break

        if not implante_encontrado:
            print("Implante no encontrado en el paciente.")

def menu():
    print("\nMenú:")
    print("1. Agregar nuevo implante")
    print("2. Eliminar implante por ID del paciente. ")
    print("3. Editar información de un implante por ID del paciente. ")
    print("4. Mostrar inventario completo")
    print("5. Salir")

lista_pacientes = []
while True:
    menu()  # Mostrar el menú
    opcion = input("Ingrese el número de la opción que desea: ")

    if opcion == "1":  # Agregar nuevo implante
        fecha = datetime.now()
        mensaje = input("Por favor ingrese los datos, si no existe escriba None")
        nombre = input("Ingrese el nombre del paciente: ")
        edad = input("Ingrese su edad: ")
        sexo = input("Ingrese 'M' o 'F': ")
        
        # Crear un nuevo paciente con los datos proporcionados
        nuevo_paciente = Paciente(nombre, edad, sexo)
        lista_pacientes.append(nuevo_paciente)
        
        fechaImplantacion = fecha
        medico = input("Nombre del médico: ")
        estado = input("Estado del implante: ")
        material = input("Material del implante: ")
        IDpaciente = input("Ingrese ID del paciente, sin puntos, comas o espacios: ")
        electrodos = input("Ingrese el número de electrodos: ")
        sistema = input("Ingrese el sistema del implante (inalámbrico o alámbrico): ")
        frecuencia_estimulacion = input("Ingrese la frecuencia de estimulación: ")
        longitud = input("Ingrese la longitud del implante: ")
        diametro = input("Ingrese el diámetro del implante: ")
        fijacion = input("Ingrese el tipo de fijación: ")
        forma = input("Ingrese la forma del implante: ")
        tamanoRodilla = input("Ingrese el tamaño de la rodilla: ")
        tamanoCadera = input("Ingrese el tamaño de la cadera: ")

        nuevo_paciente.asignar_implante(fechaImplantacion, medico, estado, material, IDpaciente, electrodos, sistema, frecuencia_estimulacion, longitud, diametro, fijacion, forma, tamanoRodilla, tamanoCadera)
        print("Implante agregado exitosamente.")
        
    elif opcion == "2":  # Eliminar implante
        id_paciente = input("Ingrese el ID del paciente del cual desea eliminar el implante: ")
        implante_eliminado = False
        for paciente in lista_pacientes:
            if paciente.IDpaciente == id_paciente:
                paciente.eliminar_implante(id_paciente)
                implante_eliminado = True
                break
        if not implante_eliminado:
            print("Paciente no encontrado.")
        
    elif opcion == "3":  # Editar información de un implante
        
        id_paciente = input("Ingrese el ID del paciente del cual desea editar el implante: ")
        implante_editado = False
        for paciente in lista_pacientes:
            if paciente.IDpaciente == id_paciente:
                # Aquí se podría implementar la lógica para solicitar los datos actualizados del implante
                # y luego llamar a la función editar_implante con los argumentos apropiados
                paciente.editar_implante(IDimplante= id_paciente, fechaImplantacion= nueva_fecha, medico= nuevo_medico, estado= nuevo_estado, material= nuevo_material)
                implante_editado = True
                break
        if not implante_editado:
            print("Implante no encontrado en el paciente.")
        pass
    elif opcion == "4":  # Mostrar inventario completo
        paciente1.mostrar_implantes()
    elif opcion == "5":  # Salir del programa
        print("¡Hasta luego!")
        break
    else:
        print("Opción inválida. Por favor, ingrese un número del 1 al 5.")



    
    

        





        


        
        
    
```
