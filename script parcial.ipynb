```python
#Punto 2
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
#Generando un n√∫mero aleatorio dentro del rango 1-5
muestras= 1000
frecuencia= 1
x=np.linspace(0, 2*np.pi,muestras)  #Tiempo
colores= ['blue', 'green', 'red', 'cyan', 'magenta', 'gray']
plt.figure(figsize=(16,14)) #crear figura donde vendr√°n las gr√°ficas
for i in range(5):
    amplitud=np.random.randint(1,6)
    y= amplitud * np.sin(2*np.pi*frecuencia* x)
    plt.plot(x, y, color=colores[i], label=f'Amplitud: {amplitud}')
    plt.xlabel("Tiempo(s)")
    plt.ylabel("Amplitud")
    plt.title(f"Gr√°fico sinusoidal con frecuencia de {frecuencia} Hz, {muestras} muestras")
    plt.legend(f"{amplitud}")
    plt.grid(True)

```
